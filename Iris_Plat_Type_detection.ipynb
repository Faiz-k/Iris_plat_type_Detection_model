```python
inputs=data.drop(['Id','Species'],'columns')
outputs=data.drop(['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],'columns')
```

```python
import sklearn 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
```

We will take K value is 13 because the formula to find k is 
> K=sqrt(No. of rows)\
> K has 150 rows root of that is 12.2 hence we will take 13

```python
model=KNeighborsClassifier(n_neighbors=13)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,y_test)
```
